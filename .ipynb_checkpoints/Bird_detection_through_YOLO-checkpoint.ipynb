## Bird detection with YOLO

Object detection is one of the most interesting applications of convolutional neural networks. The technique localizes the detected object on the image and draws a bounding box around it.

### YOLO:
YOLO ("You Only Look Once") is one of the fastest object detection algorithm. The algorithm predicts the classes and bounding boxes for multiple objects in the image in a single run of the algorithm.

Here we will use this algorithm to crop out the bounding boxes around birds in an image.

Let's first import the necessary modules.

```python
import cv2
from model.yolo_model import YOLO 
from glob import glob
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

Next let's define some functions:

Function for preprocessing the images.

```python
def process_image(img):
    """Resize, reduce and expand image.

    # Argument:
        img: original image.

    # Returns
        image: ndarray(64, 64, 3), processed image.
    """
    image = cv2.resize(img, (416, 416),
                       interpolation=cv2.INTER_CUBIC)
    image = np.array(image, dtype='float32')
    image /= 255.
    image = np.expand_dims(image, axis=0)

    return image
```

Function to read object classes from a file.

```python
def get_classes(file):
    """Get classes name.

    # Argument:
        file: classes name for database.

    # Returns
        class_names: List, classes name.

    """
    with open(file) as f:
        class_names = f.readlines()
    class_names = [c.strip() for c in class_names]

    return class_names
```

Function to crop the bounding box around the detected object with the maximum score.

```python
def crop_bird_image(image, boxes, scores, classes, all_classes):
    """crop the object box on the image.

    # Argument:
        image: original image.
        boxes: ndarray, boxes of objects.
        classes: ndarray, classes of objects.
        scores: ndarray, scores of objects.
        all_classes: all classes name.
    """
    image2=image.copy()
    bird_boxes = []
    bird_scores = []

    for box, cl, score in zip(boxes, classes, scores):
        if all_classes[cl] == 'bird':
            bird_boxes.append(box)
            bird_scores.append(score)

    if len(bird_boxes) > 0:
        ind = np.argmax(bird_scores)
        box = bird_boxes[ind]

        x, y, w, h = box
        #print(x,y,w,h)
        xm = np.floor(x + w/2 + 0.5).astype(int)
        ym = np.floor(y + h/2 + 0.5).astype(int)
        frac = 1.1
        #print(xm,ym,frac*w,frac*h)
        top = max(0, np.floor(xm - (frac*w/2) - 0.5).astype(int))
        left = max(0, np.floor(ym - (frac*h/2) - 0.5).astype(int))
        right = min(image2.shape[0], np.floor(ym + (frac*h/2) + 0.5).astype(int))
        bottom = min(image2.shape[1], np.floor(xm + (frac*w/2) + 0.5).astype(int))
        #print(top,left,bottom,right)
        image2 = image2[left:right,top:bottom]

    return image2
```

Function to draw a bounding boxes around all the detected objects in the image, also show the object class and the detection accuracy.

```python
def draw(image, boxes, scores, classes, all_classes):
    """Draw the boxes on the image.

    # Argument:
        image: original image.
        boxes: ndarray, boxes of objects.
        classes: ndarray, classes of objects.
        scores: ndarray, scores of objects.
        all_classes: all classes name.
    """
    for box, score, cl in zip(boxes, scores, classes):
        x, y, w, h = box
        xm = np.floor(x + w/2 + 0.5).astype(int)
        ym = np.floor(y + h/2 + 0.5).astype(int)
        frac=1.1
        top = max(0, np.floor(xm - (frac*w/2) - 0.5).astype(int))
        left = max(0, np.floor(ym - (frac*h/2) - 0.5).astype(int))
        right = min(image.shape[1], np.floor(xm + (frac*w/2) + 0.5).astype(int))
        bottom = min(image.shape[0], np.floor(ym + (frac*h/2) + 0.5).astype(int))

        cv2.rectangle(image, (top, left), (right, bottom), (255, 0, 0), 2)
        cv2.putText(image, '{0} {1:.2f}'.format(all_classes[cl], score),
                    (top, left - 6),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.6, (0, 0, 255), 1,
                    cv2.LINE_AA)

        print('class: {0}, score: {1:.2f}'.format(all_classes[cl], score))
        print('box coordinate x,y,w,h: {0}'.format(box))
```

The training of YOLO model takes a long time so we are going to load an existing pretrained YOLO model.

```python
yolo = YOLO(0.6, 0.5)
```

The model is trained do detect objects from a list of 80 classes with the names stored in the file `coco_classes.txt`. Let's get the classes.

```python
file = 'data/coco_classes.txt' # file containing class names for the database
                                # (80 classes for different objects including 'bird')
all_classes = get_classes(file) # get all the class names from the file
```

```python
# Reading the input image
path = '../../../images/bird.jpg' 
image1 = cv2.imread(path)
```

Running YOLO

```python
# processing the image
pimage = process_image(image1)

# running yolo to get the bounding boxes, classes and scores of the detected objects
start = time.time()
boxes, classes, scores = yolo.predict(pimage, image1.shape)
end = time.time()

print('run time: {0:.2f}s'.format(end - start))

if boxes is not None:
    # cropping out the bounding box around the object with the maximum score
    image2 = crop_bird_image(image1, boxes, scores, classes, all_classes)
    # showing the bounding box in the original image
    draw(image1, boxes, scores, classes, all_classes)
```

```python
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(5,5))
image1_m = cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
ax1.imshow(image1_m,aspect='auto')
image2_m = cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
ax2.imshow(image2_m)
```
